##### 8. Use figure() and subplot() to create a 2x2 layout: - Plot 4 different visualizations (line, bar, scatter, histogram) of relevant Titanic data.

##### 9.. Explain the difference between when to use a line plot vs a scatter plot using Titanic data as an example.

```python
 # Line Plot: Used when you want to show how something changes in order (like fare over passenger ID)
plt.figure(figsize=(6,4))
plt.plot(f['PassengerId'], f['Fare'], color='orange')
plt.title('Fare by Passenger ID')
plt.xlabel('Passenger ID')
plt.ylabel('Fare')
plt.grid()

# This line plot shows how the ticket fare changes from one passenger to the next.
```

```python
# Scatter Plot: Used when you want to see the connection between two things (like age and fare)
plt.figure(figsize=(6,4))
plt.scatter(f['Age'], f['Fare'], color='green', edgecolors='black', alpha=0.6)
plt.title('Age vs Fare')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.grid()

# This scatter plot shows if there's any pattern between age and fare.
# It helps to check if older or younger passengers paid more or less.
```

##### 10. . Create a stacked bar plot showing the number of survivors and non-survivors in each passenger class.

```python
import numpy as np
data = f.groupby(['Pclass', 'Survived']).size().unstack(fill_value=0)
         # [1, 2, 3]
not_survived = data[0].values     # Survived = 0
survived = data[1].values        # Survived = 1
x = np.arange(len(data))
labels = ['Class 1', 'Class 2', 'Class 3']
plt.xlabel('Passenger Class')
plt.ylabel('Number of Passengers')
plt.title('Survivors and Non-Survivors by Class')
# Create stacked bar chart
plt.bar(x, not_survived, color='red', label='Not Survived')
plt.bar(x, survived, bottom=not_survived, color='green', label='Survived')
```
