```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([12,23,32,37,44,49,51,56,59,64])
y = np.array([10,14,19,25,30,35,40,43,51,59])
z = np.array([70,65,61,58,50,48,42,37,28,15])
```

#### Line Plot

```python
plt.title("Line Plot")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.plot(x,y, color = 'black' , marker = 'v' ,mfc ='pink' ,ls = '--');
plt.grid()
```

```python
plt.title("Line Plot")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.plot(x,y, color = 'black' , marker = 's' ,mfc ='red' ,ls = '--', label = '2023');
plt.plot(x,z, color = 'black' , marker = 'v' ,mfc ='pink' ,ls = '--', label = '2024');
plt.grid()
plt.legend();
```

```python
plt.figure(figsize =(16,9)) ## size of the graph
plt.subplot(2,2,1)
plt.title("Line Plot")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.plot(x,y, color = 'black' , marker = 'v' ,mfc ='pink' ,ls = '--');
plt.grid()
plt.subplot(2,2,2)
plt.title("Line Plot")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.plot(x,y, color = 'black' , marker = 's' ,mfc ='red' ,ls = '--', label = '2023');
plt.plot(x,z, color = 'black' , marker = 'v' ,mfc ='pink' ,ls = '--', label = '2024');
plt.grid()
plt.legend();
```

### Bar Graph

```python
courses = ['python','dart','php','ruby','c++']
students = [56,40,30,41,65]
```

```python
plt.title('Bar Graph')
plt.xlabel('courses')
plt.ylabel('no. of students')
plt.bar(courses,students, color = 'pink', width = 0.5,edgecolor = 'black');
plt.grid()
```
