#### SVM (Support Vector Machine)

```python
# data =>letterdata.csv
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('letterdata.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
x = df.drop('letter', axis = 1)
y = df['letter']
```

```python
x
```

```python
y
```

```python
sns.countplot(x = y);
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=0,test_size =0.25)
```

```python
from sklearn.svm import SVC
```

```python
# 1. Polynomial Kernel
```

```python
svc_poly = SVC(kernel = 'poly')
```

```python
svc_poly.fit(x_train,y_train)
```

```python
y_pred = svc_poly.predict(x_test)
```

```python
from sklearn.metrics import ConfusionMatrixDisplay , accuracy_score
```

```python
ConfusionMatrixDisplay.from_predictions(y_test,y_pred)
```

```python
poly_accuracy = accuracy_score(y_test,y_pred)
poly_accuracy
```

```python
# 2. Linear Kernel
```

```python
svc_linear = SVC(kernel='linear')
```

```python
svc_linear.fit(x_train,y_train)
```

```python
y_pred = svc_linear.predict(x_test)
```

```python
ConfusionMatrixDisplay.from_predictions(y_test,y_pred)
```

```python
linear_acc = accuracy_score(y_test,y_pred)
linear_acc
```

```python
# 3. Sigmoid Kernel
```

```python
svc_sigmoid = SVC(kernel='sigmoid')
```

```python
svc_sigmoid.fit(x_train,y_train)
```

```python
y_pred = svc_sigmoid.predict(x_test)
```

```python
ConfusionMatrixDisplay.from_predictions(y_test,y_pred)
```

```python
sig_acc = accuracy_score(y_test,y_pred)
sig_acc
```

```python
# 4. RBF Kernel (Radial Basis Function)
```

```python
svc_rbf = SVC(kernel='rbf')
```

```python
svc_rbf.fit(x_train,y_train)
```

```python
y_pred = svc_rbf.predict(x_test)
```

```python
ConfusionMatrixDisplay.from_predictions(y_test,y_pred)
```

```python
rbf_acc = accuracy_score(y_test,y_pred)
rbf_acc
```
