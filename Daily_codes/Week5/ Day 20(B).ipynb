## Normalization

```python
# seperate the data
```

```python
 x = df[['Age','EstimatedSalary']]
```

```python
x
```

```python
# MIn Max Technique in normalization

#  X' =  X - min/Max - Min  

# formula to normalize value means normalize the values between [0-1]
   
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
x_scaled = scaler.fit_transform(x)
```

```python
x_scaled;
```

```python
x_scaled.max()
```

```python
x_scaled.min()
```

```python
x_scaled[:,0].max(), x_scaled[:,0].min()
```

```python
x_scaled[:,1].max(),x_scaled[:,1].min()
```

```python
plt.title("scatter plot")
plt.xlabel('salary')
plt.ylabel('Age')
plt.scatter(x_scaled[:,1],x_scaled[:,0] , c = df['Purchased'])
```
