```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
tips.shape
```

```python
# Histogram
```

```python
sns.histplot(tips['total_bill'], kde = False,bins = 15)
plt.title("Total bill distribution")
```

```python
# Kde plot
```

```python
sns.kdeplot(tips['total_bill'], fill = True)
plt.title("KDE of total bill")
```

```python
sns.histplot(tips['total_bill'], kde = True,bins = 15)
plt.title("Total bill distribution")
```

```python
# Boxplot
```

```python
sns.boxplot(tips['tip']);
```

```python
Q1 = np.percentile(tips['tip'],25)
Q3 = np.percentile(tips['tip'],75)
Q1,Q3
```

```python
IQR = Q3 - Q1
```

```python
IQR
```

```python
ub = Q3+1.5*IQR
```

```python
ub
```

```python
lb = Q1-1.5*IQR
```

```python
lb
```

```python
# Remove the outliers
x = tips[(tips['tip']>lb) & (tips['tip']<ub)]
```

```python
sns.boxplot(x['tip']);
```

# violin plot -> Kde + boxplot

```python
sns.violinplot(tips)
```

# stripplot -> Jittered scatter plot

```python
sns.stripplot(x = 'day' , y = 'total_bill', data = tips, jitter = True);
```

```python
sns.swarmplot(x = 'day' , y ='total_bill',data = tips)
```

```python
sns.scatterplot(x ='total_bill', y ='tip',data= tips,hue = 'day')
```

```python
sns.regplot(x = 'total_bill',y = 'tip',data =tips)
```
